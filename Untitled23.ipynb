```python
plt.figure(figsize=(10,5))
sns.countplot('quality',data =winequal_mod, order = winequal_mod['quality'].value_counts().index,palette="PuBuGn_d")
plt.title("quality Counts",fontsize=18,fontweight="bold")
plt.xlabel('quality')
plt.ylabel('Frequency')
plt.show()
```

```python
space = np.ones(7)/10
winequal_mod['quality'].value_counts().plot(kind='pie', explode=space, fontsize=14, autopct='%3.1f%%', wedgeprops=dict(width=0.19), 
                                       shadow=True, startangle=160, figsize=(20,20), cmap='inferno', legend=True)

plt.ylabel('quality')
plt.title('Donut Plot showing the proportion of each quality value')
```

```python
winequal_mod.plot(kind='scatter', x='quality', y='alcohol', figsize=(10, 6), color='purple', grid=False)

plt.title('Scatter plot showing the variation of quality with alcohal present in each item')
```
